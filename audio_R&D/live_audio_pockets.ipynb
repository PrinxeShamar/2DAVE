## convert text to bytes

```python
import wave
from pocketsphinx import Segmenter
import pyaudio
with wave.open("output.wav", "r") as w:
    segmenter = Segmenter(sample_rate=w.getframerate())
    for seg in segmenter.segment(w.getfp()):
        print(seg.start_time, seg.end_time)
        print(seg.pcm)
```
