```python
import numpy as np
def rms(audio_io):
    samples = np.frombuffer(audio_io, dtype=np.int16).astype(np.float64)
    rms = int(np.sqrt(np.mean(np.square(samples))))
    return rms
```

```python
fs = 16000
chunk = int(fs / 2)
recognizer = KaldiRecognizer(model, fs)
mic = pyaudio.PyAudio()
stream = mic.open(format = pyaudio.paInt16, channels = 1, rate = fs, input = True, frames_per_buffer= chunk)
stream.start_stream()
speech = b""
talking = False
counter = 0

while True:
    
    data = stream.read(chunk)

    silence_thresh = 250

    audio_rms = rms(data)

    if audio_rms > silence_thresh and not talking:
        print("Started Talking")
        talking = True

    if audio_rms <= silence_thresh and talking:
        counter += 1
        if counter > 2:
            print("Stopped Talking")
            break
    else:
        counter = 0
        

    # stream.stop_stream()
    
    text = recognizer.Result()
    print(text)
```

```python
import vosk
import pyaudio

def recognize_speech():
    fs = 16000
    chunk = int(fs / 4)
    model = vosk.Model(model_name="vosk-model-small-en-us-0.4")
    # model = vosk.Model(model_name="vosk-model-en-us-0.22-lgraph")
    recognizer = vosk.KaldiRecognizer(model, fs)

    # Initialize PyAudio
    p = pyaudio.PyAudio()

    # Open a stream using the microphone
    stream = p.open(format=pyaudio.paInt16,
                    channels=1,
                    rate=fs,
                    input=True,
                    frames_per_buffer=chunk)

    print("Listening...")

    try:
        while True:
            # Read audio data from the microphone
            data = stream.read(chunk)
            
            # Process the audio data with Vosk
            if recognizer.AcceptWaveform(data):
                result = recognizer.Result()
                print(result)
                # if "dave" in result.lower():
                #     break
            else:
                print(recognizer.PartialResult())

    except KeyboardInterrupt:
        print("Interrupted.")

    finally:
        # Close the microphone stream
        stream.stop_stream()
        stream.close()

        # Close PyAudio
        p.terminate()

if __name__ == "__main__":
    recognize_speech()
```
